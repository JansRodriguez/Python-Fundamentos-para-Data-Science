Preenchendo Valores Ausentes com Pandas

A função fillna() é usada para preencher os valores ausentes. </br>A função oferece muitas opções. </br>Podemos usar um valor específico, uma função agregada (por exemplo, média) ou o valor anterior ou seguinte.
</br>
</br>
Para ese exemplo usaremos a moda, a estatítica que representa o valor que aparece mais vezes em uma variável.

```python
import pandas as pd
```

```python
dataframe = pd.read_csv('dataset.csv')
dataframe.head()
```

```python
dataframe.isnull().sum()
```

```python
dataframe.shape
```

```python
## Extraindo a moda da Coluna com valores ausentes
moda = dataframe['Quantidade'].value_counts().index[0]
## O atributo index[0], solicita a obtenção do valor maior, no caso, que mais aparece na listagem.
moda ## Ai, nesse caso, o valor que mais se repete, vai estar armazenado na variável (no caso, moda)
```

```python
## Preencher valores ausentes (NA) com a moda
dataframe['Quantidade'].fillna(value=moda, inplace=True)
dataframe.isnull().sum()
```

```python
dataframe.head()
```

```python
## Fazendo uma Query dentro do dataframe, no modo SQL
    ## Dois modos de trazer informações de uma coluna
        ## 01. 
dataframe.Valor_Venda.describe()
```

```python
        ## 02.
dataframe['Valor_Venda'].describe()
```

```python
df2 = dataframe.query('250 < Valor_Venda < 10000')
df2['Valor_Venda'].describe()
```

```python
df3 = df2.query('Valor_Venda > 800')
df3.info()
```

```python
### Operadores Lógicos para Manipilação de Dados
df5 = dataframe[(dataframe['Segmento'] == 'Home Office') & (dataframe['Regiao'] == 'South')]
df5
# | essa barra é o OR, quando queremos uma coisa ou outra.
# != esses símbolos são não, negação de uma afirmativa
```

```python
df5.sample(5)
```

```python
## AGRUPAMENTO DE DADOS VIA GROUP BY
df6 = dataframe[['Segmento', 'Regiao', 'Valor_Venda']].groupby(['Segmento', 'Regiao']).mean()
df6
```

```python
## Agregando Funções via groupby() e agg()
df7 = dataframe[['Segmento', 'Regiao', 'Valor_Venda']].groupby(['Segmento', 'Regiao']).agg(['mean', 'std', 'count'])
    ## O groupby é por coluna não-númericas e o agg() é pela coluna númerica
df7
```

```python
## Manipulando dados via Strings
    ## Trabalhando com as funções StartWith e EndsWith
df8 = dataframe[dataframe['Segmento'].str.startswith('Con')]
df8.head()
```

```python
df8['Segmento'].value_counts()
```

```python
dataframe['Segmento'].value_counts()
```

```python
df9 = dataframe[dataframe['Segmento'].str.endswith('ice')]
df9.head()
```

```python
## Trabalhando com quebras do valor da coluna em colunas, utilizando o método SPLIT
dataframe['ID_Pedido'].head(8)
```

```python
dataframe['ID_Pedido'].str.split('-')
## O método retorna uma lista, que pode ser manipulada.
```

```python
dataframe['Ano'] = dataframe['ID_Pedido'].str.split('-').str[1]
dataframe['CodigoPais'] = dataframe['ID_Pedido'].str.split('-').str[0]
dataframe['NumeroPedido'] = dataframe['ID_Pedido'].str.split('-').str[2]
dataframe[['ID_Pedido', 'CodigoPais', 'Ano', 'NumeroPedido']].head()
```

```python
dataframe.head()
```
